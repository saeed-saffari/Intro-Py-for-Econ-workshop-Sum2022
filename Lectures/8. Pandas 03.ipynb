## Extra Work

```python
price_AAPL = data.DataReader('AAPL', 'yahoo', start, end)
```

```python
price_AAPL.head()
```

```python
price_AAPL['daily_diffrence'] = price_AAPL['Adj Close'].diff()
price_AAPL.head()
```

```python
price_AAPL['signal'] = np.where(price_AAPL['daily_diffrence'] > 0, 1, 0)
price_AAPL.head(10)
```

```python
price_AAPL['poisition'] = price_AAPL['signal'].diff()
price_AAPL.head(10)
```

```python
price_AAPL[price_AAPL['poisition'] == 1].index
```

```python
price_AAPL.loc[price_AAPL['poisition'] == 1, 'Adj Close']
```

```python
plt.figure(figsize=(15,8))

plt.plot(price_AAPL['Adj Close'], c = 'r', lw = 2)

plt.plot(price_AAPL[price_AAPL['poisition'] == 1].index,
        price_AAPL.loc[price_AAPL['poisition'] == 1, 'Adj Close'], 'o', c = 'g', markersize = 7)

plt.plot(price_AAPL[price_AAPL['poisition'] == -1].index,
        price_AAPL.loc[price_AAPL['poisition'] == -1, 'Adj Close'], 'x', c = 'k', markersize = 7)

plt.grid()
plt.show()
```
