### Task 2.3: find maximum infection rate for all of the countries. 

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.index
```

```python
country = list(corona_dataset_aggregated.index)
```

```python
max_infection = []

for c in country:
    max_infection.append(corona_dataset_aggregated.loc[c].diff().max())
    
corona_dataset_aggregated['max_infection_rate'] = max_infection
```

```python
corona_dataset_aggregated.head()
```

### Task 2.4: create a new dataframe with only needed column 

```python
corona_2 = pd.DataFrame(corona_dataset_aggregated['max_infection_rate'])
```

### Task 5.1: Plotting GDP vs maximum Infection rate

```python
x = data["Logged GDP per capita"]
y = data['max_infection_rate']

sns.scatterplot(data = data, x = x, y = np.log(y))
plt.show()
```

```python
sns.regplot(data = data, x = x, y = np.log(y))

plt.title("Logged GDP vs maximum Infection rate")

plt.grid()
plt.show()
```

### Task 5.2: Plotting Social support vs maximum Infection rate

```python
x = data["Social support"]
y = data['max_infection_rate']

sns.scatterplot(data = data, x = x, y = np.log(y))
plt.grid()
plt.show()

sns.regplot(data = data, x = x, y = np.log(y))

plt.title("Social support vs maximum Infection rate")

plt.grid()
plt.show()
```

### Task 5.3: Plotting Healthy life expectancy vs maximum Infection rate

```python
x = data["Healthy life expectancy"]
y = data['max_infection_rate']

sns.scatterplot(data = data, x = x, y = np.log(y))
plt.grid()
plt.show()

sns.regplot(data = data, x = x, y = np.log(y))

plt.title("Healthy life expectancy vs maximum Infection rate")

plt.grid()
plt.show()
```

### Task 5.4: Plotting Freedom to make life choices vs maximum Infection rate

```python
x = data["Freedom to make life choices"]
y = data['max_infection_rate']

sns.scatterplot(data = data, x = x, y = np.log(y))
plt.grid()
plt.show()

sns.regplot(data = data, x = x, y = np.log(y))

plt.title("Freedom to make life choices vs maximum Infection rate")

plt.grid()
plt.show()
```
