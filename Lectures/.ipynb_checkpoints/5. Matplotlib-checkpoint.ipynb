We'll now take an in-depth look at the Matplotlib package for visualization in Python. Matplotlib is a multi-platform data visualization library built on NumPy arrays, and designed to work with the broader SciPy stack. It was conceived by John Hunter in 2002, originally as a patch to IPython for enabling interactive MATLAB-style plotting via gnuplot from the IPython command line.

One of Matplotlib’s most important features is its ability to play well with many operating systems and graphics backends. Matplotlib supports dozens of backends and output types, which means you can count on it to work regardless of which operating system you are using or which output format you wish. This cross-platform, everything-to-everyone approach has been one of the great strengths of Matplotlib. It has led to a large user base, which in turn has led to an active developer base and Matplotlib’s powerful tools and ubiquity within the scientific Python world.

## Import
Just as we use the ``np`` shorthand for NumPy and the ``pd`` shorthand for Pandas, we will use some standard shorthands for Matplotlib imports.

The ``plt`` interface is what we will use most often, as we shall see throughout this chapter.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Specification

```python
%matplotlib inline
```

```python
x = np.linspace(-10,10,50)
y = x ** 2

plt.figure(figsize=(12,7))
plt.plot(x,y,
         c = 'r', ls = '-', lw = 3, marker = 'o', markersize = 12, markeredgecolor = 'b',
        markerfacecolor = 'm', label = "$y = x^2$")

plt.legend(fontsize = 12, loc = 'upper center')

plt.grid(c = 'k', ls = '--', alpha = 0.5)

plt.title('Function $Y = x^2$\nin range (-10,10)', fontsize = 15, fontweight = 2, fontname = 'Times New Roman', c = 'darkblue')
plt.xlabel('X range', fontsize = 12, fontweight = 2, fontname = 'Times New Roman')
plt.ylabel('Y range', fontsize = 12, fontweight = 2, fontname = 'Times New Roman')

#plt.text(-15,60, 'First text\nin figure!', fontsize = 12, fontname = 'Times New Roman')
#plt.text(-5,60, 'First text\nin figure!', fontsize = 12, fontname = 'Times New Roman')

#plt.xlim(left = -5, right =15)
#plt.ylim(20, 150)

plt.axhline(40, c = 'g', ls = '--', lw = 2)
plt.axvline(0 , c = 'b', ls = '--', lw = 2)

#plt.savefig('test_1.jpg', dpi = 700)

plt.show()
```

```python
x = np.linspace(-10,10,50)
y = x ** 2

plt.figure(figsize=(12,7))

plt.plot(x, y * 10, c = 'r', ls = '-', lw = 3, label = "$y=x^2$")
plt.plot(x, x ** 3, c = 'g', ls = '--', lw = 3, label = "$y=x^3$")
plt.plot(x, 150 * np.sin(x), c = 'b', ls = '-.', lw = 3, label = "$y=sin(x)$")

plt.legend(fontsize = 12)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,7))

plt.subplot(2,3,1)
plt.plot(x, x**2)

plt.subplot(2,3,2)
plt.plot(x, x**3)
plt.grid()

plt.subplot(2,3,3)
plt.plot(x, np.sin(x))

plt.subplot(2,3,5)
plt.plot(x, np.tan(x))

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,7))

ax.plot(x, x**2, label = 'ONE')
ax.plot(x, 50 * np.sin(x), label = 'TWO')

ax.grid()
ax.legend()

ax.set_title('Compare')
ax.set_xlabel('x range')
ax.set_ylabel('y range')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize = (12,7),sharex=True )

ax1.plot(x, x**2, label = 'ONE', c = 'r')
ax2.plot(x, 50*np.sin(x), label ='TWO', c ='b')

ax1.grid()
ax2.grid()

ax1.legend()
ax2.legend()

ax1.set_title('Compare')
#ax2.set_title('compare 2')
#ax1.set_xlabel('x range')
ax2.set_xlabel('x range')
ax1.set_ylabel('y range')
ax2.set_ylabel('y range')


plt.show()
```

## Bar

```python
plt.figure(figsize=(12,7))

plt.bar([1,3,5,7], [5,8,2,6], label = 'ONE')
plt.bar([2,4,6,8], [1,4,8,3], label = 'TWO')

plt.grid(axis = 'y', c = 'k', alpha = 0.8)
plt.legend()
plt.show()
```

## Scatter

```python
x = np.linspace(0,10, 100)
e = np.random.rand(100) * 100
y = x + e

scale = np.random.rand(100) * 800
color = np.random.rand(100)

plt.figure(figsize=(12,7))
plt.scatter(x, y, s= scale, c = color)

plt.grid()

plt.colorbar().set_label('\nColor bar', fontsize = 15)

plt.show()
```

```python
x = np.arange(100)
sc = np.arange(100) * 3
co = np.arange(100)
a = np.random.rand(100)
y = x + a * 20

plt.figure(figsize=(12,7))
plt.scatter(x,y,s = sc, c = co)


plt.grid()
plt.colorbar()
plt.show()
```

## Histogram

<img stc = https://wikimedia.org/api/rest_v1/media/math/render/svg/00cb9b2c9b866378626bcfa45c86a6de2f2b2e40 >

```python
data = np.random.normal(0,1, 1000000)

x = np.linspace(-4,4, 100)
y = 1 / np.sqrt(2 * np.pi) * np.exp(-x**2/2)

plt.figure(figsize=(12,7))
plt.hist(data, 50, density= 1)
plt.plot(x,y, lw = 3, c = 'k')

plt.show()
```
